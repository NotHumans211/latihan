#Algoritma
**Algoritma adalah serangkaian langkah-langkah terstruktur untuk menyelesaikan suatu masalah**

```python
#Algoritma Pengurutan (Sorting)

#Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
data = [64, 34, 25, 12, 22, 11, 90]
sorted_data = bubble_sort(data)
print(sorted_data)
```

```python
#Insertion Sort
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
        return arr
data = [64, 34, 25, 12, 22, 11, 90]
sorted_data = insertion_sort(data)
print("Data terurut",sorted_data)
```

#Algoritma Graph

**Breadth First Search (BFS):**

```python
from collections import deque
def bfs(graph, start):
    visited = set()
    queue = deque([start])
    while queue:
        vertex = queue.popleft()
        print(vertex, end=" ")
        for neighbor in graph[vertex]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
 graph = {
     'A': ['B', 'C'],
     'B': ['D', 'E'],
     'C': ['F'],
     'D': [],
     'E': ['F'],
     'F': []
}

bfs(graph, 'A')
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

num = 5
print("Faktorial dari", num, "adalah", factorial(num))
```

```python
# Define a Function
def divide(a, b):
    try:
        result = a / b   # Division operation
    except ZeroDivisionError:
        return "Error: Division by zero is not allowed!"
    return result

# List Operations
fruits = ["apple", "banana", "cherry"]

# Add an item to the list
fruits.append("orange")

# Remove an item from the list
fruits.remove("banana")

# Accessing items from the list
first_fruit = fruits[0]  # First item
last_fruit = fruits[-1]  # Last item

# Function Call
division_result = divide(10, 2)  # Valid division
division_by_zero = divide(10, 0)  # Invalid division

# Print Results
print(f"Fruits List: {fruits}")
print(f"First Fruit: {first_fruit}, Last Fruit: {last_fruit}")
print(f"Division Result: {division_result}")
print(f"Division by Zero: {division_by_zero}")
```
